# Part 1: Diagnosis

A coworker needs your help. They've been trying to fit a linear regression model to a new dataset, but their results have unacceptably high variance. They ask you if you can look at the dataset to try and figure out why.

What's causing the problem? Can you visualize it? Can you quantify it?

## Solution

The data is multicolinear (many of the features are stongly linearly correlated).

### Visualizations

To see linear correlations between two features, you can make a scatter plot. For example, the pandas function 'scatter_matrix' generates one plot for each pair of features:

```python
# truncate column names so labels will fit
X_truncated_names = X.rename(columns=lambda col_name: col_name[:7])

pd.plotting.scatter_matrix(X_truncated_names, figsize=(10,10))
plt.show()
```

### Quantifications

We can show the Pearson correlation coefficients to see how much each pair of features is redundant.

```python
pd.DataFrame(X).corr().style.background_gradient(cmap='coolwarm')
```

We can also use the variance inflation factor.

```python
# statsmodel can calculate VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

# but it expects each row of data for have a 1 to multiply with the first parameter
X_with_intercept = X.copy()
X_with_intercept.insert(0, 'intercept_term', 1)

VIF_scores = np.array([
    variance_inflation_factor(X_with_intercept.values, i)
    for i
    in range(X_with_intercept.shape[1])
])

pd.DataFrame({
    'feature_name': X_with_intercept.columns,
    'VIF': VIF_scores
})
```

# Part 2: Treatment

Help your coworker improve their model. Here's their original attempt:

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
original_model = LinearRegression().fit(X_train, y_train)
```

```python
original_model.score(X_test, y_test) # R-squared
```

## Solution

### Approach 1: Remove some high VIF features

Removing some of the features that are most highly correlated with the other might help. Let's remove the three features with the highest VIF scores.

```python
VIF_scores_without_intercept = VIF_scores[1:]
best_to_worst_vif_indices = np.argsort(VIF_scores_without_intercept)
```

```python
def score_after_dropping(drop_count):
    inds_of_dims_to_drop = best_to_worst_vif_indices[-drop_count:]
    X_train_dropped = np.delete(X_train, inds_of_dims_to_drop, axis=1)
    X_test_dropped = np.delete(X_test, inds_of_dims_to_drop, axis=1)
    drop_three_model = LinearRegression().fit(X_train_dropped, y_train)
    return drop_three_model.score(X_test_dropped, y_test)
```

```python
score_after_dropping(3)
```

Instead of just choosing to drop three arbitrarily, let's see when it does best.

```python
drop_results = {
    dims_remaining: score_after_dropping(X_train.shape[1] - dims_remaining)
    for dims_remaining
    in range(1, X_train.shape[1])
}

# indexing edge case
drop_results[X_train.shape[1]] = original_model.score(X_test, y_test)
```

```python
plt.scatter(drop_results.keys(), drop_results.values(), color='black')
plt.title('Dropping Features')
plt.xlabel('Dimensions remaining')
plt.ylabel('Linear Regression R')
plt.gca().invert_xaxis()
plt.show()
```

###  Approach 2: Reduce dimensionality using PCA

Instead of just removing the highest VIF features, we can use PCA to get a better trade-off between the number of features used and the information provided by them. First we'll reduce the number of dimensions by three again.

```python
from sklearn.decomposition import PCA
```

```python
def score_after_pca(pca_dims_count):
    pca = PCA(n_components=pca_dims_count)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    model = LinearRegression().fit(X_train_pca, y_train)
    return model.score(X_test_pca, y_test)
```

```python
score_after_pca(X_train.shape[1] - 3)
```

And again we can plot performance across other numbers of dimensions.

```python
pca_results = {
    pca_dim_count: score_after_pca(pca_dim_count)
    for pca_dim_count
    in range(1, X_train.shape[1] + 1)[::-1]
}
```

```python
plt.scatter(pca_results.keys(), pca_results.values(), color='black')
plt.title('PCA')
plt.xlabel('Dimensions remaining')
plt.ylabel('Linear Regression R')
plt.gca().invert_xaxis()
plt.show()
```
